```python
# Imports

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns



pd.set_option('display.max_columns', 100)
```

```python
# Load the sales prediction dataset

# Adjust this path to wherever you saved the CSV in your Drive

fpath = '/content/drive/MyDrive/AXSOSACADEMY/02-IntroML/Project1/Data/sales_predictions_2023.csv'



df = pd.read_csv(fpath)

df.head()
```

## Clean Data



We repeat the cleaning steps from Part 2 so this notebook can run standalone.

```python
# Drop duplicates (none found previously, but good practice)

df = df.drop_duplicates()



# Address missing values with placeholders

df['Item_Weight'] = df['Item_Weight'].fillna(0)

df['Outlet_Size'] = df['Outlet_Size'].fillna('MISSING')



# Fix inconsistent categories in Item_Fat_Content

fat_content_map = {

    'low fat': 'Low Fat',

    'LF': 'Low Fat',

    'reg': 'Regular'

}

df['Item_Fat_Content'] = df['Item_Fat_Content'].replace(fat_content_map)



# Confirm cleaning

print("Missing values:\n", df.isna().sum())

print("\nItem_Fat_Content categories:", df['Item_Fat_Content'].unique())
```

## Exploratory Data Analysis



For Part 3, we create exploratory visuals to help understand, explain, or model the data. This includes **at least one of each**:

- Histograms (numerical distributions)

- Boxplots (numerical statistical summaries)

- Countplots (categorical frequencies)

- A correlation heatmap

### Histograms - Distributions of Numerical Features

```python
# Identify numeric columns to visualize (exclude the target for now)

num_cols = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year']



# Plot a histogram for each numeric feature

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes = axes.flatten()



for i, col in enumerate(num_cols):

    sns.histplot(data=df, x=col, ax=axes[i], kde=True)

    axes[i].set_title(f'Distribution of {col}')



fig.tight_layout()

plt.show()
```

### Histogram - Target Variable (Item_Outlet_Sales)

```python
# Plot the distribution of the target variable

fig, ax = plt.subplots(figsize=(8, 5))

sns.histplot(data=df, x='Item_Outlet_Sales', kde=True, ax=ax)

ax.set_title('Distribution of Item_Outlet_Sales (Target)')

plt.show()
```

**Interpretation:** `Item_Outlet_Sales` is right-skewed - most products have relatively low sales, with a long tail of higher-selling products. `Item_Visibility` is also right-skewed with many products having very low display visibility.

### Boxplots - Statistical Summaries of Numerical Features

```python
# Boxplot for Item_MRP (price of the item)

fig, ax = plt.subplots(figsize=(8, 5))

sns.boxplot(data=df, x='Item_MRP', ax=ax)

ax.set_title('Boxplot of Item_MRP')

plt.show()
```

```python
# Boxplot for the target variable, Item_Outlet_Sales

fig, ax = plt.subplots(figsize=(8, 5))

sns.boxplot(data=df, x='Item_Outlet_Sales', ax=ax)

ax.set_title('Boxplot of Item_Outlet_Sales')

plt.show()
```

**Interpretation:** `Item_Outlet_Sales` has several values that appear as outliers on the high end (above ~$8,000), consistent with the right-skewed histogram seen above. `Item_MRP` does not show extreme outliers and has a roughly symmetric spread between ~$60 and ~$185 (IQR).

### Countplots - Frequency of Categorical Features

```python
# Countplot for Item_Fat_Content

fig, ax = plt.subplots(figsize=(6, 4))

sns.countplot(data=df, x='Item_Fat_Content', ax=ax)

ax.set_title('Count of Items by Item_Fat_Content')

plt.show()
```

```python
# Countplot for Outlet_Type

fig, ax = plt.subplots(figsize=(8, 5))

sns.countplot(data=df, x='Outlet_Type', ax=ax,

               order=df['Outlet_Type'].value_counts().index)

ax.set_title('Count of Items by Outlet_Type')

ax.tick_params(axis='x', rotation=20)

plt.show()
```

```python
# Countplot for Item_Type (16 categories - wider figure and horizontal bars

# for readability)

fig, ax = plt.subplots(figsize=(8, 8))

sns.countplot(data=df, y='Item_Type', ax=ax,

               order=df['Item_Type'].value_counts().index)

ax.set_title('Count of Items by Item_Type')

plt.show()
```

**Interpretation:** The vast majority of records belong to `Outlet_Type = 'Supermarket Type1'` (5,577 out of 8,523 rows), while `Item_Type` is dominated by `Fruits and Vegetables` and `Snack Foods`.

### Heatmap - Correlation Between Numerical Features

```python
# Compute the correlation matrix for numeric columns

corr = df.select_dtypes('number').corr()



# Plot the heatmap

fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=ax)

ax.set_title('Correlation Heatmap of Numerical Features')

plt.show()
```

**Interpretation:** `Item_MRP` has the strongest positive correlation with `Item_Outlet_Sales` (~0.57), suggesting that higher-priced items tend to generate higher sales revenue. Other numeric features show weak correlations with the target.

## Summary of EDA



- The target variable `Item_Outlet_Sales` is right-skewed with high-value outliers.

- `Item_MRP` (price) is the numeric feature most correlated with sales.

- The dataset is dominated by `Supermarket Type1` outlets and `Fruits and Vegetables`/`Snack Foods` item types.

- `Item_Fat_Content` originally had inconsistent labels which were consolidated into `Low Fat` and `Regular`.



**Two key visuals selected for the README:**

1. The correlation heatmap (shows `Item_MRP` as the strongest numeric predictor of sales)

2. The `Item_Outlet_Sales` histogram (shows the right-skewed distribution of the target variable)
